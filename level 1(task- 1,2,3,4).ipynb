# LEVEL 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset.csv")
```

# Level-1 (task 1) Determine the top three most common cuisines in the dataset, calculate the precentage of restaurants that serve each of the top cuisines

```python
Cuisines_count = df["Cuisines"].value_counts()
```

```python
sorted_Cuisines = Cuisines_count.sort_values(ascending=False)
```

```python
top_three_Cuisines = sorted_Cuisines.head(3)
```

```python
# Defining the colors for the bars
Colors = ['Red', 'Green', 'Blue']

# Plot the bar chart with colors
plt.bar(top_three_Cuisines.index, top_three_Cuisines.values, color=Colors)

# Add labels and title
plt.xlabel('Cuisines')
plt.ylabel('Cuisines_count')
plt.title('Top Three Cuisines')

# Show the chart
plt.show()
```

```python
total_restaurants = len(df)
```

```python
precentage_pre_Cuisine = (top_three_Cuisines / total_restaurants) * 100
```

```python
print(precentage_pre_Cuisine)
```

```python
plt.bar(precentage_pre_Cuisine.index, precentage_pre_Cuisine.values, color=Colors)
plt.xlabel('Cuisines')
plt.ylabel('Cuisines_count')
plt.title('Top Three Cuisines')
plt.show()
```

# Level-1 (task-2) identify the city with the highest number of restaurants in dataset, Calculate the average rating for restaurants in each city and determine the cithy with the highest average rating

```python
restaurant_count = df.groupby("City")["Restaurant Name"].count().reset_index()
```

```python
city_with_highest_restaurants = restaurant_count[restaurant_count['Restaurant Name'] == restaurant_count['Restaurant Name'].max()]['City'].values[0]
```

```python
print("The city with the highest number of restaurants is:",city_with_highest_restaurants)
```

```python
grouped_by_city = df.groupby('City')
```

```python
average_ratings = grouped_by_city['Aggregate rating'].mean()
```

```python
print(average_ratings)
```

```python
highest_rated_city = average_ratings.idxmax()
```

```python
highest_average_rating = average_ratings.max()
```

```python
print(f"The city with the highest average rating is {highest_rated_city} with an average rating of {highest_average_rating}.")
```

# Level-1 (task-3) create a histogram or bar chart to visualize the distribution of price ranges among the restaurants and calculate the precentage of restaurants in each price rane category

```python
grouped_by_price = df.groupby('Price range')
price_range_counts = grouped_by_price['Restaurant Name'].count()
```

```python
plt.bar(price_range_counts.index, price_range_counts.values)
plt.xlabel('Price range')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of Price Range Among Restaurants')
plt.show()
```

```python
percentage_per_price_range = (price_range_counts /total_restaurants) * 100
```

```python
for price_range, percentage in percentage_per_price_range.items():
    print(f"Price Range {price_range}: {percentage:.2f}%")
```

# Level-1 (task-4) determine the precentage of restaurants that offer online delivery and compare the average rating of restaurants with and without online delivery

```python
restaurants_with_delivery = (df['Has Online delivery'] == "Yes").sum()
```

```python
percentage_with_delivery = (restaurants_with_delivery / total_restaurants) * 100
```

```python
print(f"The percentage of restaurants that offer online delivery is: {percentage_with_delivery:.2f}%")
```

```python
Average_rating_with_online_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
Average_rating_without_online_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()
print(Average_rating_with_online_delivery)
print(Average_rating_without_online_delivery)
```

```python
colours = ["pink","yellow"]
labels = ['With Online Delivery', 'Without Online Delivery']
Average_ratings = [Average_rating_with_online_delivery, Average_rating_without_online_delivery]

plt.bar(labels, Average_ratings,color=colours)
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.title('Comparison of Average Ratings')
plt.show()
```
